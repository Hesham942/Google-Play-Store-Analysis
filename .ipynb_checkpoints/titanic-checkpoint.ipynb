```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[['Age', 'Fare']])

data['Age_scaled'] = scaled_data[:, 0]
data['Fare_scaled'] = scaled_data[:, 1]
```

```python
from sklearn.model_selection import train_test_split

X = data[['Age_scaled', 'Fare_scaled', 'Sex']]
y = data['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Assuming your CSV file is named 'titanic.csv'
titanic_df = pd.read_csv('tested.csv')
```

```python

# Check basic statistics
print(titanic_df.describe())


```

```python
# Handle missing values (e.g., fill or drop)
titanic_df.fillna(0, inplace=True)  # Replace 'value' with the appropriate strategy

# Drop unnecessary columns
titanic_df = titanic_df.drop(['column1', 'column2'], axis=1)
```

```python
# Example: Bar chart of survival counts
titanic_df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Example: Survival rate by passenger class
titanic_df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
