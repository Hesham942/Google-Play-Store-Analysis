```python
import pandas as pd
import psycopg2
```

```python
data1=pd.read_csv('anime.csv',index_col=False)
data2=pd.read_csv('rating.csv',index_col=False)
```

```python
df=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
merge1 = pd.merge(df,df2, on="anime_id")
merge1
```

```python
l=merge1.sort_values(by=['anime_id'])
```

```python
l
```

```python
l.rename(columns={'name':'anime_name', 'rating_x':'anime_rating','rating_y':'user_ratings'},inplace=True)
```

```python
l
```

```python
anime_state = l.drop(["anime_id","genre", "type", "episodes",'anime_rating','members','user_id'], axis=1)
anime_state
```

```python
anime_avg_ratings=anime_state.groupby(['anime_name']).mean().sort_values(['user_ratings'],ascending=False)
anime_avg_rating=anime_avg_ratings.rename(columns={'user_ratings': 'avg_rating'})
```

```python
anime_avg_rating
```

```python
anime_avg_rating.to_csv('anime_avg_rating.csv')
```

```python
import psycopg2
import pandas as pd

def connect():
    try:
        conn = psycopg2.connect(
            dbname="postgres",
            user="postgres",
            password="root",
            host="127.0.0.1",
            port="5432"
        )
        return conn
    except psycopg2.Error as e:
        print("Unable to connect to the database.")
        print(e)
        return None

def create_table(conn):
    try:
        cur = conn.cursor()
        cur.execute("""
            CREATE TABLE IF NOT EXISTS anime (
                anime_id INT PRIMARY KEY NOT NULL,
                anime_name TEXT,
                genre TEXT,
                type TEXT,
                episodes INT,
                anime_rating DOUBLE PRECISION,
                members INT,
                user_id INT,
                user_ratings INT
            );
        """)
        conn.commit()
        cur.close()
    except psycopg2.Error as e:
        print("Error creating table.")
        print(e)

# Function to insert data
def insert_data(conn, data):
    try:
        cur = conn.cursor()
        # Insert data into the table
        cur.execute("""
            INSERT INTO anime (
                anime_id,
                anime_name,
                type,
                genre,
                episodes,
                anime_rating,
                members,
                user_id,
                user_ratings
            ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s);
        """, data)
        conn.commit()
        cur.close()
    except psycopg2.Error as e:
        print("Error inserting data.")
        print(e)

# Main function
def main():
    # Load data from CSV file
    df = p # Replace "your_data.csv" with your CSV file path
    
    # Connect to PostgreSQL
    conn = connect()
    if conn is None:
        return
    
    # Create table
    create_table(conn)

    # Insert data into the table
    for index, row in df.iterrows():
        data = (row['anime_id'], row['anime_name'], row['type'], row['genre'], row['episodes'], row['anime_rating'], row['members'], row['user_id'], row['user_ratings'])
        insert_data(conn, data)

    # Close connection
    conn.close()

if __name__ == "__main__":
    main()
```
